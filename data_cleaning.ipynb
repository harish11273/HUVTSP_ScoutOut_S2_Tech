Importing Pandas, NumPy, and Dataset

```python
import pandas as pd
import numpy as np
df = pd.read_csv('C:/Users/hj920/ScoutOut Code!/Dataset/realtor-data.zip.csv')
df.head()
```

Basic Data Cleaning (Columns, NaN's, Duplicates)

```python
df.drop(columns = ["brokered_by" , "status" , "street" , "city" , "state"] , inplace = True)
df.dropna(inplace = True)
df.drop_duplicates(inplace = True)
df.head()
```

Converting dates to datetime and organizing by date in ascending order

```python
df['prev_sold_date'] = pd.to_datetime(df['prev_sold_date'])
df = df.sort_values('prev_sold_date')
df = df.reset_index(drop=True)
df.head()
```

```python
df.info()
df.describe()
df.corr()["price"]
df.corr()["prev_sold_date"]
```

```python
df.to_csv('C:/Users/hj920/ScoutOut Code!/Dataset/cleaned_realtor-data_7-31.csv')
```

Wiping all values before 1/1/1990

```python
df.shape
df.info()
df.head()
```

```python
dfr = df.drop(df[df['prev_sold_date'] < '1990-01-01'].index)
dfr = dfr.reset_index(drop=True)
dfr.shape
dfr.info()
dfr.describe()
dfr.corr()["price"]
dfr.corr()["prev_sold_date"]
dfr.head()
```
